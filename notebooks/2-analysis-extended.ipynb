### Hypotesene som undersøkes er følgende:

**Kunnskapsnivå** :<br>
H0: Chat-GPT 4.0 genererer svar på kliniske spørsmål på _samme_ kunnskapsnivå, sammenlignet svar fra utdannede og lisensierte leger (MDs).<br>
H1: Chat-GPT 4.0 genererer svar på kliniske spørsmål med _forskjellig_ kunnskapsnivå sammenlignet med leger (trolig _lavere_ kunnskapsnivå).

**Empati** :<br>
H0: Chat-GPT 4.0 genererer svar på kliniske spørsmål med _samme_ grad av empati, sammenlignet med svar fra leger.<br>
H1: Chat-GPT 4.0 genererer svar på kliniske spørsmål med _ulik_ grad av empati sammenlignet med leger (trolig _mindre_ grad av empati).

**Hjelpsomhet** :<br>
H0: Chat-GPT 4.0 genererer _samme_ grad av hjelpsomme svar på kliniske spørsmål, sammenlignet med leger.<br>
H1: Chat-GPT 4.0 genererer _forskjellig_ grad av hjelpsomme svar på kliniske spørsmål sammenlignet med leger (trolig _mindre_ hjelpsomhet).

**Respondent kategori** :<br>
H0: leger og medisinstudenter med lisens vurderer Chat-GPT 4.0 sine svar på kliniske spørsmål som _like_ kunnskapsrike som hva andre som jobber/ studerer eller har bakgrunn i helsevesenet gjør i vurderingen av Chat-GPT 4.0 sine svar mhp. kunnskap.<br>
H1: leger og medisinstudenter med lisens vurderer Chat-GPT 4.0 sine svar på kliniske spørsmå til å ha _ulik_ grad av kunnskap sammenliknet med hva andre som jobber/ studerer eller har bakgrunn i helsevesenet gjør.

````python
def wrap_text(df, column, i):
    # Get the text
    text = df[column].iloc[i]

    # Split the text into words
    words = text.split()

    # Group the words into chunks of 10
    chunks = [' '.join(words[i:i+10]) for i in range(0, len(words), 10)]

    # Join the chunks with line breaks
    wrapped_text = '\n'.join(chunks)
    

    if column == 'tekst':
        print(f"Spørsmål i=#{i} (spørsmålid:{df['spørsmålid'].iloc[i]}, vurderingid:{df['vurderingid'].iloc[i]}, respondentid:{df['respondentid'].iloc[i]}):\n{wrapped_text}")
    elif column == 'svartekst':
        if df['chatgpt'].iloc[i] == False:
            print(f"Svartekst (lege) #{i} (svarid:{df['svarid'].iloc[i]}, vurderingid:{df['vurderingid'].iloc[i]}, respondentid:{df['respondentid'].iloc[i]}):\n{wrapped_text}")
        else:
            print(f"Svartekst (chatgpt) #{i} (svarid:{df['svarid'].iloc[i]}, vurderingid:{df['vurderingid'].iloc[i]}, respondentid:{df['respondentid'].iloc[i]}):\n{wrapped_text}")
    else:
        print(f'Unknown column: {column}')
````

````python
def wrap_text_spmid(spm_id):
    # Get the text
    df_spm = df2[df2['spørsmålid'] == spm_id]
    text = df_spm['tekst'].iloc[0]
    
    # Split the text into words
    words = text.split()

    # Group the words into chunks of 10
    chunks = [' '.join(words[i:i+10]) for i in range(0, len(words), 10)]

    # Join the chunks with line breaks
    wrapped_text = '\n'.join(chunks)
    
    print(f"Spørsmål (spørsmålid:{df_spm['spørsmålid'].iloc[0]}, vurderingid:{df_spm['vurderingid'].iloc[0]}, respondentid:{df_spm['respondentid'].iloc[0]}):\n{wrapped_text}")
````

````python
def wrap_text_svarid(svar_id):
    # Get the text
    df_svar = df2[df2['svarid'] == svar_id]
    text = df_svar['svartekst'].iloc[0]

    # Split the text into words
    words = text.split()

    # Group the words into chunks of 10
    chunks = [' '.join(words[i:i+10]) for i in range(0, len(words), 10)]

    # Join the chunks with line breaks
    wrapped_text = '\n'.join(chunks)
    
    if df_svar['chatgpt'].iloc[0] == False:
        print(f"Svartekst (lege) (svarid:{df_svar['svarid'].iloc[0]}, vurderingid:{df_svar['vurderingid'].iloc[0]}, respondentid:{df_svar['respondentid'].iloc[0]}):\n{wrapped_text}")
    else:
        print(f"Svartekst (chatgpt) (svarid:{df_svar['svarid'].iloc[0]}, vurderingid:{df_svar['vurderingid'].iloc[0]}, respondentid:{df_svar['respondentid'].iloc[0]}):\n{wrapped_text}")
````

Feature importance refers to techniques that assign a score to input features based on how useful they are at predicting a target variable. There are many ways to calculate feature importance, and the importance of a feature depends on the model and the specific dataset.

In general, a higher feature importance score means that the feature is more important for the prediction. This could mean that changes in the feature's values have a larger effect on the predicted output, or that the feature helps to reduce the model's error.

However, feature importance scores do not tell you the direction of the relationship between the feature and the target (whether the relationship is positive or negative), and they do not tell you about the interactions between features.

Also, keep in mind that feature importance scores are relative: a feature with a low importance score could still be important if all features have low scores, and a feature with a high importance score might not be that important if all features have high scores.

Finally, remember that feature importance scores are based on the specific model and dataset, so they might not generalize to other models or datasets.

SHAP (SHapley Additive exPlanations) is a unified measure of feature importance that assigns each feature an importance value for a particular prediction. Its use in explaining the importance of features in a model for making predictions can be summarized as follows:

1. **Global Interpretability**: The overall importance of each feature can be measured by taking the mean absolute SHAP value over the entire dataset. Features with larger mean absolute SHAP values are more important in the model.

2. **Local Interpretability**: For a single prediction, the SHAP value of each feature represents how much each feature contributed to the difference between the prediction and the baseline prediction. Positive SHAP values increase the prediction from the baseline, while negative SHAP values decrease it.

3. **Consistency**: If a model changes so that it relies more on a particular feature, then the SHAP value for that feature should also increase.

4. **High-Dimensional Feature Interactions**: SHAP values can capture interaction effects between features. If a feature has a high SHAP value, it means that it is important either individually or as part of an interaction.

By using SHAP values, we can gain a deeper understanding of the model's decision-making process, which can help in validating or improving the model.

Morris Sensitivity Analysis, also known as Morris Method or Elementary Effects Method, is a global sensitivity analysis method used to screen for important input factors that have an effect on the output of a model. It's particularly useful when dealing with models that have a large number of input parameters.

The method works by calculating elementary effects, which measure the change in the output of a model resulting from a small change in an input parameter. The process is repeated many times, each time with a different set of randomly selected input parameters.

The sensitivity of the model output to a particular input parameter is then determined by the mean and standard deviation of the elementary effects for that parameter. The mean measures the overall influence of the parameter on the output, and the standard deviation measures the interaction and non-linear effects of the parameter.

Here's a step-by-step breakdown of the Morris Method:

1. Define a grid of possible values for each input parameter.
2. Randomly select a point in the grid and compute the model output.
3. For each input parameter, move along the grid by one step, compute the model output again, and calculate the elementary effect.
4. Repeat steps 2 and 3 many times, each time with a different randomly selected starting point.
5. For each input parameter, calculate the mean and standard deviation of the elementary effects.

The parameters with the highest mean and standard deviation are considered the most influential on the model output.

The Mann-Whitney U test is a non-parametric statistical significance test for determining whether two independent samples were drawn from a population with the same distribution.

Mathematically, the Mann-Whitney U test works as follows:

1. Combine all the values from the two samples into one set and sort them in ascending order.
2. Assign ranks to these values. The smallest value gets a rank of 1, the second smallest a rank of 2, and so on. In case of ties, assign to each tied value the average of the ranks they would have received if they were not tied.
3. Sum the ranks for the values from each of the two samples separately. Let's call these sums R1 and R2.
4. Calculate U for each sample using the formula U = n1*n2 + n*(n+1)/2 - R, where n is the size of the sample and R is the sum of the ranks in that sample. There will be two U values, U1 and U2.
5. The smaller of U1 and U2 is the Mann-Whitney U statistic.

The null hypothesis (H0) for the test is that the distributions of both populations are equal. If the p-value is less than the chosen significance level (typically 0.05), the null hypothesis is rejected and it can be concluded that there is a significant difference between the two populations.

In Python, you can use the `mannwhitneyu` function from the `scipy.stats` module to perform the Mann-Whitney U test. This function returns two values: the U statistic and the p-value. Here's an example of how to interpret the results:



````python
from scipy.stats import mannwhitneyu

# Assuming 'sample1' and 'sample2' are your two samples
sample1, sample2 = df_GPT['kunnskap'], df_human['kunnskap']

# Perform the Mann-Whitney U test
U, p = mannwhitneyu(sample1, sample2)

print('U statistic:', U)
print('p-value:', p)
````



If the p-value is less than 0.05, you can reject the null hypothesis and conclude that there is a significant difference between the distributions of the two samples. If the p-value is greater than 0.05, you cannot reject the null hypothesis and you cannot conclude that there is a significant difference between the distributions of the two samples.

The Mann-Whitney U test, also known as the Wilcoxon rank-sum test, is a non-parametric test that compares two unpaired groups of data. It's used to determine whether there is a significant difference between the two groups. It's a non-parametric alternative to the independent t-test and can be used when the assumption of normality is not met.

Here's how it works:

1. **Ranking**: Combine all values from both groups into a single set and rank them in ascending order. If there are ties (equal values), assign each a rank equal to the average of the ranks they would have occupied if they were not tied.

2. **Sum of Ranks**: Calculate the sum of the ranks for each group separately.

3. **U Statistic**: Calculate the U statistic for each group. The U statistic for a group is the sum of the ranks in that group minus the theoretical minimum this sum could have (which is `(n*(n+1))/2` where `n` is the size of the group). The smaller U statistic of the two is the one used for the test.

4. **Significance**: Compare the calculated U statistic to a critical value from the U distribution (which is tabulated and depends on the sizes of the two groups). If the calculated U is less than or equal to the critical value, the null hypothesis (that the two groups are samples from populations with the same distribution) can be rejected.

In Python, you can use the `mannwhitneyu` function from the `scipy.stats` module to perform the Mann-Whitney U test. This function returns two values: the U statistic and the p-value (which is calculated using an approximation that is valid for large sample sizes). If the p-value is less than the chosen significance level (typically 0.05), the null hypothesis is rejected.

````python
# How large is the difference in the median scores of "kunnskap" for GPT and human (excluding "Vet ikke")?

median_GPT = df_GPT_kunnskap_excl6['kunnskap'].median()
median_human = df_human_kunnskap_excl6['kunnskap'].median()
diff = median_GPT - median_human
print(f'Median score for kunnskap for GPT: {median_GPT} (excluding "Vet ikke")')
print(f'Median score for kunnskap for human: {median_human} (excluding "Vet ikke")')
print(f'Difference in median scores: {diff}')
````

````python
# How large is the difference in the mean scores of kunnskap for GPT and human?

mean_GPT = df_GPT_kunnskap_excl6['kunnskap'].mean()
mean_human = df_human_kunnskap_excl6['kunnskap'].mean()

std_GPT = df_GPT['kunnskap'].std()
std_human = df_human['kunnskap'].std()

diff = mean_GPT - mean_human
print(f'Mean score for kunnskap for GPT: {mean_GPT:.2f} ± {std_GPT:.2f} (n={len(df_GPT_kunnskap_excl6)}) (excluding "Vet ikke")')
print(f'Mean score for kunnskap for human: {mean_human:.2f} ± {std_human:.2f} (n={len(df_human_kunnskap_excl6)}) (excluding "Vet ikke")')
print(f'Difference in mean scores: {diff:.2f}')
````
